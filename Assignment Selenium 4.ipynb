#### Q1.Scrape the details of most viewed videos on YouTube from Wikipedia: Url= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:

A).Rank

B) Name

C) Artist

D) Upload date

E) Views

```python
def scrap_most_view_vidios(path,web_link):
    driver = webdriver.Chrome(path)
    driver.get(web_link)
    time.sleep(2)
    values={
        "Rank":[],
        "Name":[],
        "Artist":[],
        "Upload Date":[],
        "Views":[]
    }
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr"):
        values['Rank'].append(i.find_element_by_xpath('td[1]').text)
        values['Name'].append(re.sub("\[.*?\]","",i.find_element_by_xpath('td[2]').text.replace('"',"")))
        values['Artist'].append(i.find_element_by_xpath('td[3]').text)
        values['Views'].append(i.find_element_by_xpath('td[4]').text)
        values['Upload Date'].append(i.find_element_by_xpath('td[5]').text)
    return values

data=scrap_most_view_vidios("/home/pavankumar/Downloads/chromedriver_linux64/chromedriver","https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
pd.DataFrame(data)
```

#### Q2. Scrape the details team India’s international fixtures from bcci.tv. Url = https://www.bcci.tv/. You need to find following details:

A) Match title (I.e. 1 st ODI)

B) Series

C) Place

D) Date

E) Time

Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
def scrap_indian_cricket_fixtures(path,web_link):
    driver = webdriver.Chrome(path)
    driver.get(web_link)
    driver.get(driver.find_element_by_xpath("//a[@class='navigation__link navigation__link--in-drop-down']").get_attribute('href'))
    time.sleep(2)
    values={
        "Match title":[],
        "Series":[],
        "Place":[],
        "Date":[],
        "Time":[]
    }
    for title,date_time,series in zip(driver.find_elements_by_xpath("//div[@class='fixture__description u-unskewed-text']"),driver.find_elements_by_xpath("//div[@class='fixture__datetime desktop-only']"),driver.find_elements_by_xpath("//div[@class='fixture__teams']")):
        values['Match title'].append(title.text.split("\n")[0])
        values['Place'].append(title.text.split("\n")[1])
        date=date_time.text.split("\n")[:3]
        values['Date'].append('-'.join(date))
        values['Time'].append(date_time.text.split("\n")[-1])
        values['Series'].append(series.text.replace("\n"," "))
    return values
        
data=scrap_indian_cricket_fixtures("/home/pavankumar/Downloads/chromedriver_linux64/chromedriver","https://www.bcci.tv/")
pd.DataFrame(data)
```

#### Q3. Scrape the details of selenium exception from guru99.com. Url = https://www.guru99.com/ You need to find following details:

A) Name

B) Description

Note: - From guru99 home page you have to reach to selenium exception handling page through code.

```python
def scrap_selenium_exceptions(path,web_link):
    driver = webdriver.Chrome(path)
    driver.get(web_link)
    time.sleep(2)
    driver.find_element_by_id('java_technologies').find_element_by_xpath("li[3]").click()
    driver.find_element_by_xpath("//a[@title='Selenium Exception Handling (Common Exceptions List)']").click()
    values={
        "Name":[],
        "Description":[]}
    for i in driver.find_elements_by_xpath("//table[@class='table table-striped']/tbody/tr"):
        try:
            i.find_element_by_xpath("td/strong").text
        except NoSuchElementException as e:
            values["Name"].append(i.text.split(" ")[0])
            values['Description'].append(' '.join(i.text.split(" ")[1:]))
    return values
data=scrap_selenium_exceptions("/home/pavankumar/Downloads/chromedriver_linux64/chromedriver","https://www.guru99.com/")
pd.DataFrame(data)
```

#### Q4.Scrape the details of State-wise GDP of India from statisticstime.com. Url = http://statisticstimes.com/ You have to find following details:

A) Rank

B) State

C) GSDP(18-19)

D) GSDP(17-18)

E) Share(2017)

F) GDP($ billion)

Note: - From statisticstimes home page you have to reach to economy page through code.

```python
def scrap_gdp_state_india(path,web_link):
    driver = webdriver.Chrome(path)
    driver.get(web_link)
    time.sleep(1)
    driver.get(driver.find_element_by_xpath("//div[@class='navbar']/div[2]/div/a[3]").get_attribute('href'))
    driver.find_element_by_xpath("//div[@style='float:left;background-color:seashell;width:400px;height:800px;']/ul/li/a").click()
    values={
        "Rank":[],
        "State":[],
        "GSDP(18-19)":[],
        "GSDP(17-18)":[],
        "Share(2017)":[],
        "GDP($ billion)":[]
    }
    for i in driver.find_elements_by_xpath("//div[@class='dataTables_wrapper']/table/tbody/tr"):
        values['Rank'].append(i.find_element_by_xpath('td[1]').text)
        values['State'].append(i.find_element_by_xpath('td[2]').text)
        values['GSDP(18-19)'].append(i.find_element_by_xpath('td[3]').text)
        values['GSDP(17-18)'].append(i.find_element_by_xpath('td[4]').text)
        values['Share(2017)'].append(i.find_element_by_xpath('td[5]').text)
        values['GDP($ billion)'].append(i.find_element_by_xpath('td[6]').text)
    return values
data=scrap_gdp_state_india("/home/pavankumar/Downloads/chromedriver_linux64/chromedriver","http://statisticstimes.com/")
pd.DataFrame(data)
```

#### Q5. Scrape the details of trending repositories on Github.com. Url = https://github.com/ You have to find the following details:

A) Repository title

B) Repository description

C) Contributors count

D) Language used

Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
def scrap_github_data(path,web_link):
    driver = webdriver.Chrome(path)
    driver.get(web_link)
    time.sleep(1)
    driver.get(driver.find_element_by_xpath("//ul[@class='list-style-none mb-3']/li[3]/a").get_attribute("href"))
    values={
        "Repository title":[],
        "Repository description":[],
        "Contributors count":[],
        "Language used":[]}
    for title,desc,count,lang in zip(driver.find_elements_by_xpath("//span[@class='text-normal']"), driver.find_elements_by_xpath("//p[@class='col-9 color-text-secondary my-1 pr-4']"), driver.find_elements_by_xpath("//div[@class='f6 color-text-secondary mt-2']/a[2]"),driver.find_elements_by_xpath("//span[@itemprop='programmingLanguage']")):
        values['Repository title'].append(title.text.split()[0])
        values['Repository description'].append(desc.text)
        values['Contributors count'].append(count.text)
        values['Language used'].append(lang.text)
    return values
data=scrap_github_data("/home/pavankumar/Downloads/chromedriver_linux64/chromedriver","https://github.com/")
pd.DataFrame(data)
```

A) Song name

B) Artist name

C) Last week rank

D) Peak rank

Note: - From the home page you have to click on the charts option then hot 100-page link through code. 

```python
def scrap_top_100_songs_on_billiboard(path,web_link):
    driver = webdriver.Chrome(path)
    driver.get(web_link)
    driver.get(driver.find_element_by_xpath("//a[@class='header__submenu__list__link']").get_attribute('href'))
    data=[]
    data.append(["Rank","Position","Song name","Artist name","Last week rank","Peak rank","Weeks on board"])
    for i in driver.find_elements_by_xpath("//li[@class='chart-list__element display--flex']"):
        data.append(i.text.split("\n"))
    return data
data=scrap_top_100_songs_on_billiboard("/home/pavankumar/Downloads/chromedriver_linux64/chromedriver","https://www.billboard.com/")
pd.DataFrame(data[1:],columns=data[0])
```

#### Q7. Scrape the details of Data science recruiters from naukri.com. Url = https://www.naukri.com/ You have to find the following details:

A) Name

B) Designation

C) Company

D) Skills they hire for

E) Location

Note: - From naukri.com homepage click on the recruiters option and the on the search pane type Data science and
click on search. All this should be done through code

```python
def scrap_job_data_sci_nkry(path,web_link):
    driver = webdriver.Chrome(path)
    driver.get(web_link)
    driver.get(driver.find_element_by_xpath("//a[@title='Search Recruiters']").get_attribute('href'))
    driver.find_element_by_xpath("//input[@class='sugInp']").send_keys("Data Science")
    driver.find_element_by_id("qsbFormBtn").click()
    data=[]
    data.append(["Name","Designation","Company","Location","Skills they hire for"])
    for i in driver.find_elements_by_xpath("//div[@class='recInfo']"):
        row=i.text.split("\n")
        row.pop(4)
        data.append(row[:5])
    return data
data=scrap_job_data_sci_nkry("/home/pavankumar/Downloads/chromedriver_linux64/chromedriver","https://www.naukri.com/")
pd.DataFrame(data[1:],columns=data[0])
```

#### Q8. Scrape the details of Highest selling novels. Url = https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare/ You have to find the following details:

B) Author name

C) Volumes sold

D) Publisher

E) Genre

```python
def scrap_high_selling_novek(path,web_link):
    driver = webdriver.Chrome(path)
    driver.get(web_link)
    data=[]
    row_lis=[]
    data.append(driver.find_element_by_xpath("//table[@class='in-article sortable']/thead").text.split("\n"))
    for i in driver.find_elements_by_xpath("//table[@class='in-article sortable']/tbody/tr"):
        row_lis.append(i.find_element_by_xpath("td[1]").text)
        row_lis.append(i.find_element_by_xpath("td[2]").text)
        row_lis.append(i.find_element_by_xpath("td[3]").text)
        row_lis.append(i.find_element_by_xpath("td[4]").text)
        row_lis.append(i.find_element_by_xpath("td[5]").text)
        row_lis.append(i.find_element_by_xpath("td[6]").text)
        data.append(row_lis)
        row_lis=[]
    return data
data=scrap_high_selling_novek("/home/pavankumar/Downloads/chromedriver_linux64/chromedriver","https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare/")
pd.DataFrame(data[1:],columns=data[0])
```
