#### Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

```python
driver = webdriver.Chrome('/home/pavankumar/Downloads/chromedriver_linux64/chromedriver')
driver.get('https://www.naukri.com/')
search_box=driver.find_element_by_id('qsb-keyword-sugg')
search_box.send_keys('Data Analyst')
location_name=driver.find_element_by_id("qsb-location-sugg")
location_name.send_keys('Bangalore')
driver.find_element_by_xpath("//div[@class='search-btn']/button").click()
with open('All Job list from Noukri web site.csv','w') as f: 
    writer = csv.writer(f) 
    writer.writerow(['Job Titles', 'Company Name', 'Experiance','Job Locations']) 
    for n in range(0,50):
        for i in driver.find_elements_by_xpath("//div[@class='jobTupleHeader']"):
            if len(i.text.split('\n'))==5:
                row_lis=i.text.split('\n')
                row_lis.pop(3)
                writer.writerow(row_lis)
            else:
                row_lis=i.text.split('\n')
                row_lis.pop(2)
                row_lis.pop(2)
                row_lis.pop(3)
                writer.writerow(row_lis)
        time.sleep(5)
        driver.get(driver.find_element_by_xpath("//a[@class='fright fs14 btn-secondary br2']").get_attribute('href'))
f.close()
```

#### Q5: Write a python program to scrape the salary data for Data Scientist designation in Noida location. You have to scrape Company name, Number of salaries, Average salary, Min salary, Max Salary. 

```python
driver = webdriver.Chrome('/home/pavankumar/Downloads/chromedriver_linux64/chromedriver')
driver.get('https://www.glassdoor.co.in/Salaries/index.htm')
driver.find_element_by_id("KeywordSearch").send_keys('Data Scientist')
loc_box=driver.find_element_by_id("LocationSearch")
loc_box.clear()
loc_box.send_keys('Noida India')
driver.find_element_by_id('HeroSearchButton').click()
time.sleep(3)
with open('All Salaries from Glassdoor web site.csv','w') as f:
    header=['Company Name', 'Average Salary','Minimum Salary','Maximum Salary','No of Salary']
    writer=csv.DictWriter(f, fieldnames = header)
    writer.writeheader()
    for i,j,k in zip(driver.find_elements_by_xpath("//div[@class='d-flex']/div[2]"),driver.find_elements_by_xpath("//div[@class='col-2 d-none d-md-flex flex-row justify-content-end']"),driver.find_elements_by_xpath("//div[@class='common__RangeBarStyle__values d-flex justify-content-between ']")):
        writer.writerow({
            'Company Name':i.text.split("\n")[1],
            'Average Salary':j.text.split('₹ ')[1].replace("\n",' '),
            'Minimum Salary':k.text.split('\n')[0],
            'Maximum Salary':k.text.split('\n')[1],
            'No of Salary':i.text.split("\n")[2]
        })

f.close()
```

#### Q6 : Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
1. Brand
2. Product Description
3. Price
4. Discount % 

```python
driver = webdriver.Chrome('/home/pavankumar/Downloads/chromedriver_linux64/chromedriver')
driver.get('https://www.flipkart.com/')
search_box=driver.find_element_by_xpath("//input[@class='_3704LK']")
search_box.send_keys('sunglasses')
driver.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']").click()
driver.find_element_by_xpath("//button[@class='L0Z3Pu']").click()
with open('All Sunglasses from Flipkart web site.csv','w') as f: 
    writer = csv.writer(f) 
    writer.writerow(['Brand Name', 'Product Description','Discount %','Price'])
    for n in range(1,5):
        for i in driver.find_elements_by_xpath("//div[@class='_2B099V']"):
            lis=i.text.split('\n')
            if len(lis)==3:
                discount=i.text.split('\n')[2][-7:]
                price = i.text.split('\n')[2].split('₹')[1]
                lis.pop()
                lis=lis+[discount,price]
            else:
                lis.pop()
                discount=i.text.split('\n')[2][-7:]
                price = i.text.split('\n')[2].split('₹')[1]
                lis.pop()
                lis=lis+[discount,price]
            time.sleep(1)
            writer.writerow(lis)
        time.sleep(2)
        if driver.find_element_by_xpath("//div[@class='_2MImiq']/nav/a[11]").text=="NEXT":
            driver.find_element_by_xpath("//div[@class='_2MImiq']/nav/a[11]").click()
        else:
            driver.find_element_by_xpath("//div[@class='_2MImiq']/nav/a[12]").click()
        time.sleep(1)
f.close()
```

#### Q10: Go to webpage https://www.amazon.in/ Enter “Laptop” in the search field and then click the search icon. Then set CPU Type filter to “Intel Core i7” and “Intel Core i9” .After setting the filters scrape first 10 laptops data. You have to scrape 3 attributes for each laptop:
1. title
2. Ratings
3. Price

```python
driver = webdriver.Chrome('/home/pavankumar/Downloads/chromedriver_linux64/chromedriver')
driver.get('https://www.amazon.in/')
search_box=driver.find_element_by_id("twotabsearchtextbox")
search_box.send_keys('lap tops')
driver.find_element_by_id('nav-search-submit-button').click()
with open('All Laptops from Amazon web site.csv','w') as f: 
    header = ['Title', 'Ratings','Price']
    writer = csv.DictWriter(f, fieldnames = header) 
    writer.writeheader() 
    for i in range(0,20):
        try:
            if driver.find_element_by_xpath("//ul[@aria-labelledby='p_n_feature_thirteen_browse-bin-title']/li[{}]".format(2+i)).text=='Intel Core i7':
                driver.find_element_by_xpath("//ul[@aria-labelledby='p_n_feature_thirteen_browse-bin-title']/li[{}]/span".format(2+i)).click()
                break
        except:
            break
    for j in range(0,20):
        try:
            if driver.find_element_by_xpath("//ul[@aria-labelledby='p_n_feature_thirteen_browse-bin-title']/li[{}]".format(2+i)).text=='Intel Core i9':
                driver.find_element_by_xpath("//ul[@aria-labelledby='p_n_feature_thirteen_browse-bin-title']/li[{}]/span".format(2+i)).click()
                break
        except:
            break
    time.sleep(3)
    for n in range(1,3):
        title=[]
        rating=[]
        price=[]
        for i,j,k in zip(driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']"),driver.find_elements_by_xpath("//span[@class='a-price']"),driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']")):
            title.append(k.text)
            price.append(j.text)
            try:
                driver1 = webdriver.Chrome('/home/pavankumar/Downloads/chromedriver_linux64/chromedriver')
                time.sleep(1)
                driver1.get(i.get_attribute("href"))
                time.sleep(1)
                rating.append(driver1.find_element_by_xpath("//span[@class='a-size-medium a-color-base']").text)
                driver1.close()
            except:
                rating.append("NONE")
                driver1.close()
        for t,r,p in zip(title,rating,price):
            writer.writerow({
                "Title":t,
                'Ratings':r,
                'Price':p,
            })
        time.sleep(2)
        driver.find_element_by_xpath("//li[@class='a-last']").click()
        time.sleep(4)
f.close()
```
